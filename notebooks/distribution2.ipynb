```python
import numpy as np
import matplotlib.pyplot as plt

from scipy.stats import weibull_min
```

## Fixed move order
L,U,D,R

```python
data = np.load('../distribution/fixed.npy')
max_tile = np.sort(data[:, 0])
score = np.sort(data[:, 1])
moves = np.sort(data[:, 2])
```

```python
plt.figure(figsize=(12, 4))
plt.subplot(131)
plt.plot(max_tile, '.')
plt.title('Max tile log 2')
plt.subplot(132)
plt.plot(score, '.')
plt.title('Score')
plt.subplot(133)
plt.plot(moves, '.')
plt.title('Moves')
plt.show()
```

Not able to reach 1024 in 5000 games

```python
plt.figure(figsize=(8, 4))
plt.subplot(121)
plt.hist(score, bins=50)
plt.title('Score')
plt.subplot(122)
plt.hist(moves, bins=50)
plt.title('Moves')
plt.show()
```

We can model the "time-to-failure" using a Weibull distribution.

```python
dist = score
fit = weibull_min.fit(dist, loc=0, scale=dist.mean())
print(fit)
plt.hist(dist, bins=50, density=True)
xx = np.linspace(0, dist.max())
plt.plot(xx, weibull_min(*fit).pdf(xx))
plt.show()
```

```python
dist = moves
fit = weibull_min.fit(dist, loc=14, scale=np.median(moves))
# absolute minimum moves should be 14
print(fit)
plt.hist(dist, bins=50, density=True)
xx = np.linspace(0, dist.max())
plt.plot(xx, weibull_min(*fit).pdf(xx))
plt.show()
```

# Percentages

```python
def plot_max(max_tiles):
    percents, _, _ = plt.hist(max_tiles, bins=range(16), density=True)
    plt.show()

    max_tile = int(np.amax(max_tiles))
    for i in range(4):
        print(f'{percents[max_tile-i]}% reached {max_tile-i} ({2**(max_tile-i)})')
```

```python
def get_max_tiles(x):
    tiles = []
    for _ in range(16):
        tiles.append(x & 0xF)
        x >>= 4
    return np.amax(tiles, axis=0)
```

### Fixed

```python
data = np.load('../distribution/fixed.npy')
max_tiles = np.sort(data[:, 0])

plot_max(max_tiles)
```
